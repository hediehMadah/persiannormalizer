```python
import pickle as pkl
import pandas as pd
import re
import numpy as np
addr = 'D:/SecondTerm/NLP/tweets.pkl'
with open(addr,'rb') as myData:  
  tweets = pkl.load(myData)
tweets
```

```python
df = pd.DataFrame(tweets,columns=['متن'])
df.head(20)
```

```python
from hazm import *
normalizer = Normalizer()
```

```python
def normal(txt):
    return normalizer.normalize(txt)
```

```python
df['متن'] = df['متن'].apply(normal)
df.head(20)
```

```python
def remove_punctuation(txt):
    return re.sub(r'[^آ-ی]+', ' ', txt)
```

```python
df['متن'] = df['متن'].apply(remove_punctuation)
```

```python
df.head(20)
```

```python
import persian #before pip install persian
def convert_ar_chars(txt):
    s = ''
    for word in txt:
        s = s + persian.convert_ar_characters(word) 
    return s
```

```python
df['متن'] = df['متن'].apply(convert_ar_chars)
df.head(20)
```
